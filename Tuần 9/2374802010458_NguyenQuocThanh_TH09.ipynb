# 8. Phân tích cảm xúc người dùng

```python
nltk.download('movie_reviews')
```

```python
import random
reviews = nltk.corpus.movie_reviews
documents = [(list(reviews.words(fileid)), category)
for category in reviews.categories()
for fileid in reviews.fileids(category)]
random.shuffle(documents)
first_review = ' '.join(documents[0][0])
print(first_review)
```

```python
documents[0][1]
```

```python
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
# Chỉ lấy 2000 từ phổ biến nhất để làm đặc trưng
word_features = list(all_words.keys())[:2000]
```

```python
def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['{}'.format(word)] = (word in document_words)
    return features

# Chạy lại việc trích xuất đặc trưng với tập từ vựng đã rút gọn
featuresets = [(document_features(d, word_features), c) for (d, c) in documents]
print(len(featuresets))
```

```python
# Chia dữ liệu và huấn luyện
train_set, test_set = featuresets[500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)

# Kiểm tra độ chính xác
print("Accuracy:", nltk.classify.accuracy(classifier, test_set))
classifier.show_most_informative_features(15)
```

```python
# 4. Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.
from nltk.tokenize import word_tokenize
# Văn bản mẫu
text = "This is a sample sentence showing how to remove stopwords using NLTK library."
# Tải dữ liệu cần thiết
nltk.download('punkt')
# Thiết lập ngôn ngữ
stop_words = set(stopwords.words('english'))
# Tokenize và lọc
word_tokens = word_tokenize(text)
filtered_text = [w for w in word_tokens if w.lower() not in stop_words]
print("Văn bản gốc:", text)
print("Văn bản sau khi loại bỏ stopword:", filtered_text)
```

```python
# 5. Viết chương trình Python với thư viện NLTK bỏ qua các stopword từ danh sách các stopword.
from nltk.corpus import stopwords

# Lấy danh sách stopword tiếng Anh mặc định
stop_words = set(stopwords.words('english'))
print(f"Số lượng stopword ban đầu: {len(stop_words)}")

# Danh sách các từ bạn muốn 'bỏ qua' (không coi là stopword nữa)
ignore_stopwords = {'not', 'no', 'but', 'against'}

# Loại bỏ các từ này khỏi set stop_words
custom_stop_words = stop_words - ignore_stopwords

print(f"Số lượng stopword sau khi lọc: {len(custom_stop_words)}")
print(f"Các từ đã được giữ lại (không còn là stopword): {ignore_stopwords}")
```

```python
# 6. Viết một chương trình Python với thư viện NLTK để tìm định nghĩa và ví dụ của một từ đã cho bằng WordNet
from nltk.corpus import wordnet

# Tải dữ liệu wordnet nếu chưa có
nltk.download('wordnet')
nltk.download('omw-1.4')

word = 'bank'
synsets = wordnet.synsets(word)

print(f"Từ: {word}")
if synsets:
    for i, syn in enumerate(synsets[:3]): # Hiển thị 3 nghĩa đầu tiên
        print(f"\nNghĩa {i+1}: {syn.name()}")
        print(f"- Định nghĩa: {syn.definition()}")
        if syn.examples():
            print(f"- Ví dụ: {syn.examples()[0]}")
else:
    print("Không tìm thấy từ này trong WordNet.")
```

```python
# 7. Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó.
import nltk
from nltk.corpus import wordnet

# Tải dữ liệu wordnet nếu chưa có
nltk.download('wordnet')

word = 'good'
synonyms = []
antonyms = []

for syn in wordnet.synsets(word):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print(f"Từ: {word}")
print(f"- Từ đồng nghĩa: {set(synonyms)}")
print(f"- Từ trái nghĩa: {set(antonyms)}")
```

```python
# 8. Viết chương trình Python với thư viện NLTK để có cái nhìn tổng quan về bộ tag, chi tiết của
# một tag cụ thể trong bộ tag và chi tiết về một số bộ tag liên quan, sử dụng biểu thức chính quy.
import nltk
import re

# Tải dữ liệu tagsets và tagsets_json nếu chưa có
nltk.download('tagsets')
nltk.download('tagsets_json')

# 1. Cái nhìn tổng quan về bộ nhãn (tagset)
print("--- Tổng quan về bộ nhãn UPENN ---")
nltk.help.upenn_tagset()

# 2. Chi tiết của một tag cụ thể
target_tag = 'NN'
print(f"\n--- Chi tiết của tag: {target_tag} ---")
nltk.help.upenn_tagset(target_tag)

# 3. Sử dụng biểu thức chính quy để tìm các bộ nhãn liên quan
regex_pattern = r'VB.*'
print(f"\n--- Các nhãn liên quan đến pattern: {regex_pattern} ---")

tags = ['CC', 'CD', 'DT', 'EX', 'FW', 'IN', 'JJ', 'JJR', 'JJS', 'LS', 'MD', 'NN', 'NNS', 'NNP', 'NNPS', 'PDT', 'POS', 'PRP', 'PRP$', 'RB', 'RBR', 'RBS', 'RP', 'SYM', 'TO', 'UH', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'WDT', 'WP', 'WP$', 'WRB']

related_tags = [t for t in tags if re.match(regex_pattern, t)]

for tag in related_tags:
    print(f"\nChi tiết cho nhãn {tag}:")
    nltk.help.upenn_tagset(tag)
```

```python
# 9. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã cho
import nltk
from nltk.corpus import wordnet

# Tải dữ liệu wordnet nếu chưa có
nltk.download('wordnet')

def compare_nouns(noun1, noun2):
    # Lấy tập hợp các nghĩa (synsets) cho mỗi từ
    syns1 = wordnet.synsets(noun1, pos=wordnet.NOUN)
    syns2 = wordnet.synsets(noun2, pos=wordnet.NOUN)

    if not syns1 or not syns2:
        return f"Không tìm thấy dữ liệu cho một trong hai từ: {noun1}, {noun2}"

    # So sánh nghĩa đầu tiên (phổ biến nhất) của mỗi từ
    s1 = syns1[0]
    s2 = syns2[0]

    similarity = s1.path_similarity(s2)

    print(f"Từ 1: {noun1} ({s1.definition()})")
    print(f"Từ 2: {noun2} ({s2.definition()})")
    print(f"Độ tương đồng (Path Similarity): {similarity:.4f}")

# Ví dụ so sánh
compare_nouns('dog', 'cat')
print('-' * 30)
compare_nouns('dog', 'car')
```

```python
# 10. Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai động từ đã cho
import nltk
from nltk.corpus import wordnet

# Tải dữ liệu wordnet nếu chưa có
nltk.download('wordnet')

def compare_verbs(verb1, verb2):
    # Lấy tập hợp các nghĩa (synsets) cho mỗi từ với pos là VERB
    syns1 = wordnet.synsets(verb1, pos=wordnet.VERB)
    syns2 = wordnet.synsets(verb2, pos=wordnet.VERB)

    if not syns1 or not syns2:
        return f"Không tìm thấy dữ liệu cho một trong hai động từ: {verb1}, {verb2}"

    # So sánh nghĩa đầu tiên của mỗi từ
    v1 = syns1[0]
    v2 = syns2[0]

    similarity = v1.path_similarity(v2)

    print(f"Động từ 1: {verb1} ({v1.definition()})")
    print(f"Động từ 2: {verb2} ({v2.definition()})")
    print(f"Độ tương đồng (Path Similarity): {similarity:.4f}")

# Ví dụ so sánh
print("--- So sánh động từ ---")
compare_verbs('run', 'walk')
print('-' * 30)
compare_verbs('run', 'eat')
```

```python
# 12. Viết chương trình Python với thư viện NLTK để in 15 kết hợp ngẫu nhiên đầu tiên được gắn nhãn nam và được gắn nhãn tên nữ từ kho tên.
import random

# Create a list of tuples (name, gender)
names_labeled = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

# Shuffle the list to get random combinations
random.shuffle(names_labeled)

# Print the first 15 random labeled names
print("15 random labeled names from the corpus:")
for name, gender in names_labeled[:15]:
    print(f"{name}: {gender}")
```
